# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import inspect
from dateutil.relativedelta import relativedelta
from flask import Flask, jsonify
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
base = automap_base()
# reflect the tables
base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
base.classes.keys()
```

```python
# Save references to each table
measurements = base.classes.measurement
stations = base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Perform a query to retrieve the data and precipitation scores
##remember: q = some_session.query(User, Address); use this to perfrom a query
precipitation = session.query(measurements.date, measurements.prcp).\
    filter(measurements.date >= '2016-08-23',measurements.date <= '2017-08-23' ).\
    order_by(measurements.date).all()
print(precipitation)
```

```python
# Design a query to show how many stations are available in this dataset?
##refer to query design for dates
##stations = session.query(measurements.station).order_by(measurements.station.desc()).all()
##print(stations)
```

```python
##number of stations that are available in this dataset
stations = session.query(func.count(stations.station)).all()
print(stations)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
##use the group by funtion instead of filter 
stations = session.query(measurements.station, func.count(measurements.station)).\
    group_by(measurements.station).order_by(func.count(measurements.station).desc()).all()
print(stations)
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
df_stations = pd.DataFrame(stations, columns=['station', 'tobs'])
print(df_stations)
```

```python
##get the temperatures together
temps = session.query(measurements.tobs, func.count(measurements.tobs)).\
    group_by(measurements.tobs).order_by(func.count(measurements.tobs).desc()).all()
print(temps)
```

```python
##get the temperature variations (min., max., and avg.)
#temps_variation = session.query(measurements.tobs, func.min(measurements.tobs)).\
#                     (measurements.tobs, func.max(measurements.tobs)).\
#                     (measurements.tobs, func.avg(measurements.tobs)).\
#                     order_by(measurements.tobs).all()
#    print(temps_variation)
##clean up code

temps_variation = session.query(func.min(measurements.tobs), func.max(measurements.tobs), func.avg(measurements.tobs)).all()
print(temps_variation)
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
## create query with everything 
erryting = session.query(measurements.tobs).\
    filter(measurements.station == 'USC00519281').\
    filter(measurements.date >= year_ago).\
    order_by(measurements.date.desc()).all()
erryting_df = pd.DataFrame(erryting)
erryting_df.head()
```

```python
##make a histogram 
##use bins for the last 12 months
plt.hist(erryting_df['tobs'], bins=12)
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Station Analysis 2016-2017")
plt.legend(["tobs"])
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```
